```python
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
import seaborn as sns

# Buat directed graph
G = nx.DiGraph()

# Tambah edge: username  in_reply_to_screen_name
for _, row in df.iterrows():
    source = row['username']
    target = row['in_reply_to_screen_name']
    if pd.notnull(source) and pd.notnull(target):
        G.add_edge(source, target)

# Tampilkan info dasar graph
print("Jumlah node:", G.number_of_nodes())
print("Jumlah edge:", G.number_of_edges())

# Hitung derajat (jumlah koneksi) tiap node
degree_dict = dict(G.degree())
degree_df = pd.DataFrame(list(degree_dict.items()), columns=['User', 'Degree'])

# Visualisasi distribusi derajat node
plt.figure(figsize=(10, 6))
sns.histplot(degree_df['Degree'], bins=10, kde=True)
plt.title("Distribusi Derajat Node (Degree Distribution)")
plt.xlabel("Degree (Jumlah koneksi)")
plt.ylabel("Jumlah Node")
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
import seaborn as sns

# Buat directed graph
G = nx.DiGraph()

# Tambahkan edge dari username ke in_reply_to_screen_name
for _, row in df.iterrows():
    source = row['username']
    target = row['in_reply_to_screen_name']
    if pd.notnull(source) and pd.notnull(target):
        G.add_edge(source, target)

# Daftar seed nodes berdasarkan eigenvector centrality
seed_nodes = ['BANGSAygSUJUD', 'Kaum__Jelata', 'nuka_minona', 'Mawar___Merah', '1onto']

# Tampilkan hasil
print("Seed nodes berdasarkan Eigenvector Centrality:")
for user in seed_nodes:
    print(f"- {user}")
```

```python
import random  # Pastikan sudah import ini

# === Fungsi simulasi ICM ===
def run_icm(graph, seeds, activation_prob=0.9):
    active = set(seeds)
    newly_active = set(seeds)
    all_activated = set(seeds)

    while newly_active:
        next_active = set()
        for node in newly_active:
            neighbors = set(graph.successors(node)) - all_activated
            for neighbor in neighbors:
                if random.random() <= activation_prob:
                    next_active.add(neighbor)
        newly_active = next_active
        all_activated.update(newly_active)

    return all_activated

# === Seed nodes berdasarkan eigenvector centrality ===
seed_nodes = ['BANGSAygSUJUD', 'Kaum__Jelata', 'nuka_minona', 'Mawar___Merah', '1onto']

# === Jalankan simulasi ===
hasil_aktivasi = run_icm(G, seed_nodes, activation_prob=0.9)

# === Output hasil ===
print("Jumlah seed nodes:", len(seed_nodes))
print("Total node yang teraktivasi:", len(hasil_aktivasi))
print("Daftar node yang teraktivasi:")
for node in hasil_aktivasi:
    print("-", node)
```

```python
import matplotlib.pyplot as plt

# Fungsi ICM tetap pakai yang sebelumnya
def run_icm(graph, seeds, activation_prob=0.1):
    active = set(seeds)
    newly_active = set(seeds)
    all_activated = set(seeds)

    while newly_active:
        next_active = set()
        for node in newly_active:
            neighbors = set(graph.successors(node)) - all_activated
            for neighbor in neighbors:
                if random.random() <= activation_prob:
                    next_active.add(neighbor)
        newly_active = next_active
        all_activated.update(newly_active)

    return all_activated

# === Uji beberapa nilai probabilitas ===
probabilitas = [0.1, 0.3, 0.5, 0.7, 0.9]
num_simulasi = 100
rata_rata_hasil = []

for prob in probabilitas:
    total = 0
    for _ in range(num_simulasi):
        hasil = run_icm(G, seed_nodes, activation_prob=prob)
        total += len(hasil)
    rata_rata = total / num_simulasi
    rata_rata_hasil.append(rata_rata)

# === Visualisasi hasil ===
plt.figure(figsize=(8, 6))
plt.plot(probabilitas, rata_rata_hasil, marker='o', linestyle='-', color='blue')
plt.title("Pengaruh Probabilitas Aktivasi terhadap Penyebaran ICM")
plt.xlabel("Probabilitas Aktivasi")
plt.ylabel("Rata-rata Jumlah Node Teraktivasi")
plt.grid(True)
plt.xticks(probabilitas)
plt.show()
```

```python
import matplotlib.pyplot as plt

# === Jalankan 1x simulasi untuk divisualisasikan ===
hasil_aktivasi = run_icm(G, seed_nodes, 0.9)

# === Atur warna node ===
node_colors = []
for node in G.nodes():
    if node in seed_nodes:
        node_colors.append('red')       # Seed nodes
    elif node in hasil_aktivasi:
        node_colors.append('green')     # Node yang teraktivasi
    else:
        node_colors.append('black')     # Node yang tidak aktif

# === Posisi node (layout) ===
pos = nx.spring_layout(G, seed=42)  # seed biar layout konsisten

# === Gambar grafik ===
plt.figure(figsize=(12, 10))
nx.draw_networkx_nodes(G, pos, node_size=50, node_color=node_colors, alpha=0.8)
nx.draw_networkx_edges(G, pos, alpha=0.3, arrows=True, width=0.5)
plt.title("Visualisasi Penyebaran Informasi (ICM)", fontsize=14)
plt.axis('off')
plt.show()
```
